### Graph Cut Segmentation using GrabCut and a pretrained Faster R-CNN detector

#### Import Liibraries

```python
import os
from pathlib import Path
import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import torch
from torchvision import transforms
from torchvision.models.detection import fasterrcnn_resnet50_fpn
```

```python
# Device (CPU by default; if CUDA enabled PyTorch is there, set device to 'cuda')
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

#### Helper Functions

#### Display images

```python
def show_images_grid(images, titles=None, figsize=(12,6), cmap=None):
    n = len(images)
    cols = min(3, n)
    rows = (n + cols - 1) // cols
    plt.figure(figsize=figsize)
    for i, img in enumerate(images):
        plt.subplot(rows, cols, i+1)
        if img is None:
            plt.axis('off')
            continue
        if img.ndim == 2:
            plt.imshow(img, cmap=cmap)
        else:
            plt.imshow(img)
        if titles:
            plt.title(titles[i])
        plt.axis('off')
    plt.tight_layout()
```

#### Load image with OpenCV (BGR -> RGB)

```python
def load_img_rgb(path):
    bgr = cv2.imread(str(path))
    if bgr is None:
        raise FileNotFoundError(f"Image not found: {path}")
    rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
    return rgb
```

#### Apply GrabCut given rectangle

```python
def run_grabcut_rgb(img_rgb, rect, iterations=5):
    # img_rgb: HxWx3 (uint8, RGB)
    img_bgr = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2BGR)
    mask = np.zeros(img_bgr.shape[:2], np.uint8)  # 0 background, 1 foreground - used by grabCut as cv2.GC_BGD etc
    bgdModel = np.zeros((1,65), np.float64)
    fgdModel = np.zeros((1,65), np.float64)
    x, y, w, h = rect
    rect_cv = (int(x), int(y), int(w), int(h))
    cv2.grabCut(img_bgr, mask, rect_cv, bgdModel, fgdModel, iterations, cv2.GC_INIT_WITH_RECT)
    # mask values: 0,2 are background; 1,3 are foreground
    mask2 = np.where((mask == 2) | (mask == 0), 0, 1).astype('uint8')
    foreground = img_rgb * mask2[:, :, np.newaxis]
    # overlay: blend original with red mask
    overlay = img_rgb.copy()
    alpha = 0.6
    red = np.zeros_like(img_rgb)
    red[:, :, 0] = 255
    overlay = np.where(mask2[:, :, np.newaxis]==1, (img_rgb * (1-alpha) + red * alpha).astype(np.uint8), img_rgb)
    return mask2, foreground, overlay
```

#### Expand bounding box slightly while keeping within image

```python
def expand_box(box, img_shape, margin_ratio=0.15):
    x1, y1, x2, y2 = box
    h, w = img_shape[:2]
    bw = x2 - x1
    bh = y2 - y1
    margin = int(max(bw, bh) * margin_ratio)
    nx1 = max(0, x1 - margin)
    ny1 = max(0, y1 - margin)
    nx2 = min(w-1, x2 + margin)
    ny2 = min(h-1, y2 + margin)
    return nx1, ny1, nx2, ny2
```

#### Load images

```python
image_files = ["asm-1.jpg", "asm-2.jpg"]
images = []
for fn in image_files:
    p = Path(fn)
    if not p.exists():
        print(f"Warning: {fn} not found in working directory ({os.getcwd()}). Please add your images and re-run.")
    images.append(load_img_rgb(p) if p.exists() else None)
```

#### Import pretrained model

```python
print("Loading pretrained Faster R-CNN (this may take a moment)...")

model = fasterrcnn_resnet50_fpn(pretrained=True, progress=True)
model.to(device)
model.eval()
```

```python
# COCO person class index is 1
PERSON_CLASS = 1
```

```python
# transform images
transform = transforms.Compose([
    transforms.ToTensor(),
])
```

#### # Run detector and get bounding boxes for 'person'

```python
detections = []

for i, img_rgb in enumerate(images):
    if img_rgb is None:
        detections.append(None)
        continue
    img_tensor = transform(img_rgb).to(device)
    with torch.no_grad():
        preds = model([img_tensor])[0]
    boxes = preds['boxes'].cpu().numpy()
    labels = preds['labels'].cpu().numpy()
    scores = preds['scores'].cpu().numpy()
    # filter person detections with reasonable score
    person_inds = np.where(labels == PERSON_CLASS)[0]
    person_boxes = []
    for idx in person_inds:
        person_boxes.append((boxes[idx], scores[idx]))
    # sort by score descending
    person_boxes = sorted(person_boxes, key=lambda x: x[1], reverse=True)
    if len(person_boxes) == 0:
        print(f"No person detected in {image_files[i]}")
        detections.append(None)
    else:
        # take top detection
        top_box, top_score = person_boxes[0]
        x1, y1, x2, y2 = top_box.astype(int)
        detections.append({
            'box': (x1, y1, x2, y2),
            'score': float(top_score),
            'all_persons': person_boxes
        })
        print(f"Image {image_files[i]}: detected person with score {top_score:.3f} -> box {(x1,y1,x2,y2)}")

        # === Display bounding box result before GrabCut ===
        boxed_img = img_rgb.copy()
        cv2.rectangle(boxed_img, (x1, y1), (x2, y2), (0, 255, 0), 3)
        plt.figure(figsize=(6,6))
        plt.imshow(boxed_img)
        plt.title(f"Detected person in {image_files[i]} (score={top_score:.2f})")
        plt.axis("off")
        plt.show()
```

#### Prepare rectangles for GrabCut and run for iterations 1,3,5

```python
results = []

iter_list = [1, 3, 5]
for i, img_rgb in enumerate(images):
    if img_rgb is None or detections[i] is None:
        results.append(None)
        continue
    det = detections[i]
    x1,y1,x2,y2 = det['box']
    ex1, ey1, ex2, ey2 = expand_box((x1,y1,x2,y2), img_rgb.shape, margin_ratio=0.2)
    rect = (ex1, ey1, ex2-ex1, ey2-ey1)  # x, y, w, h for OpenCV
    per_image_res = {'rect': rect, 'runs': {}}
    for iters in iter_list:
        mask2, foreground, overlay = run_grabcut_rgb(img_rgb, rect, iterations=iters)
        # quantitative metrics: foreground pixel count, foreground fraction, contour length
        fg_pixels = int(mask2.sum())
        total_pixels = mask2.size
        fg_fraction = fg_pixels / total_pixels
        # contour length
        contours, _ = cv2.findContours(mask2.astype(np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        contour_length = sum(float(cv2.arcLength(cnt, True)) for cnt in contours)
        per_image_res['runs'][iters] = {
            'mask': mask2,
            'foreground': foreground,
            'overlay': overlay,
            'fg_pixels': fg_pixels,
            'fg_fraction': fg_fraction,
            'contour_length': contour_length
        }
    results.append(per_image_res)
```

#### Quantitative comparison between iterations

```python
for i, res in enumerate(results):
    if res is None:
        continue
    print(f"\nMetrics for {image_files[i]}")
    prev_mask = None
    for iters in iter_list:
        r = res['runs'][iters]
        print(f" Iter {iters}: fg_pixels={r['fg_pixels']}, fg_fraction={r['fg_fraction']:.4f}, contour_length={r['contour_length']:.1f}")
        if prev_mask is not None:
            # Jaccard index between prev and current masks
            curr_mask = r['mask']
            inter = np.logical_and(prev_mask, curr_mask).sum()
            union = np.logical_or(prev_mask, curr_mask).sum()
            jaccard = inter / union if union>0 else 0.0
            print(f"    Jaccard vs previous: {jaccard:.4f} (intersection {inter}, union {union})")
        prev_mask = r['mask']
```

#### Visualization per image

```python
# --- Visualization per image ---
for i, img_rgb in enumerate(images):
    if img_rgb is None or results[i] is None:
        continue
    res = results[i]
    rect = res['rect']
    x, y, w, h = rect

    print(f"\nResults for {image_files[i]}")

    # Draw bounding box on original image
    boxed = img_rgb.copy()
    cv2.rectangle(boxed, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Create user mask (rectangle region)
    user_mask = np.zeros_like(img_rgb)
    user_mask[y:y + h, x:x + w, :] = img_rgb[y:y + h, x:x + w, :]

    # Prepare rows of images
    # Row 1: Original and user mask
    row1_imgs = [boxed, user_mask]
    row1_titles = ["Original (with box)", "User mask (rectangle)"]

    # Row 2: Foreground-only images for 1, 3, 5 iterations
    row2_imgs = []
    row2_titles = []
    for iters in iter_list:
        row2_imgs.append(res['runs'][iters]['foreground'])
        row2_titles.append(f"Foreground (iter={iters})")

    # Row 3: Overlay images for 1, 3, 5 iterations
    row3_imgs = []
    row3_titles = []
    for iters in iter_list:
        row3_imgs.append(res['runs'][iters]['overlay'])
        row3_titles.append(f"Overlay (iter={iters})")

    # Display all rows
    fig, axes = plt.subplots(3, max(len(row2_imgs), 2), figsize=(14, 10))

    # Row 1
    for j in range(2):
        axes[0, j].imshow(row1_imgs[j])
        axes[0, j].set_title(row1_titles[j])
        axes[0, j].axis('off')
    for j in range(2, max(len(row2_imgs), 2)):
        axes[0, j].axis('off')

    # Row 2 - foregrounds
    for j in range(len(row2_imgs)):
        axes[1, j].imshow(row2_imgs[j])
        axes[1, j].set_title(row2_titles[j])
        axes[1, j].axis('off')

    # Row 3 - overlays
    for j in range(len(row3_imgs)):
        axes[2, j].imshow(row3_imgs[j])
        axes[2, j].set_title(row3_titles[j])
        axes[2, j].axis('off')

    plt.tight_layout()
    plt.show()
```

#### Save outputs to disk for inspection

```python
out_dir = Path('grabcut_outputs')
out_dir.mkdir(exist_ok=True)
for i, img_rgb in enumerate(images):
    if img_rgb is None or results[i] is None:
        continue
    base = Path(image_files[i]).stem
    res = results[i]
    for iters, run in res['runs'].items():
        Image.fromarray(run['foreground']).save(out_dir / f"{base}_foreground_iter{iters}.png")
        Image.fromarray(run['overlay']).save(out_dir / f"{base}_overlay_iter{iters}.png")
    # also save the rectangle visualization
    x,y,w,h = res['rect']
    boxed = img_rgb.copy()
    cv2.rectangle(boxed, (x,y), (x+w, y+h), (0,255,0), 2)
    Image.fromarray(boxed).save(out_dir / f"{base}_boxed.png")

print("\nSaved outputs to:", out_dir.resolve())
print("Done.")
```

#### Discussion

The goal is to see how the number of iterations affects the segmentation quality.

GrabCut is an iterative algorithm: it refines its segmentation mask step by step, updating the model for foreground and background colors using Gaussian Mixture Models (GMMs).
So, with more iterations, it should theoretically produce a more accurate segmentation  up to a point.

 What to Look For:
Aspect	What Happens	How to Describe It
Foreground clarity	At low iterations (1), some parts of the person may be missing or roughly cut.	At 1 iteration, the persons outline is incomplete, and the segmentation includes background noise.

Edge sharpness	Edges become more precise with more iterations.	At 3 iterations, boundaries between the person and background are cleaner.

Background leakage	Early iterations may include unwanted background pixels.	At 1 iteration, background pixels near the persons arm are still classified as foreground.

Color consistency	Foreground color model improves with iterations.	At 5 iterations, the foreground appears more uniformly colored, indicating stable segmentation.

Stability	After enough iterations (usually 5+), results dont change much.	The 5-iteration result looks similar to 3 iterations, suggesting convergence.
 Example sentence:

Qualitatively, the segmentation improves from 1 to 3 iterations  the persons silhouette becomes more accurate and less noisy. By 5 iterations, the improvement is minor, indicating convergence.


Quantitative:

Quantitative Comparison (Numerical / Metric Analysis)

This is based on numerical measurements your code already calculates:

Foreground pixel count

Foreground fraction (proportion of image labeled as foreground)

Contour length

Jaccard index (overlap between successive masks)


| Metric                     | Meaning                                      | How to Interpret Changes                                                                                         |
| -------------------------- | -------------------------------------------- | ---------------------------------------------------------------------------------------------------------------- |
| **Foreground pixel count** | Total number of pixels labeled as foreground | If this decreases across iterations, GrabCut is refining and removing background noise.                          |
| **Foreground fraction**    | Ratio of foreground to total pixels          | Helps measure how much of the image GrabCut considers person. Stable values indicate convergence.              |
| **Contour length**         | Total edge length of the segmentation mask   | Very high = noisy edges, unstable mask. Decreasing = smoother and more consistent boundaries.                    |
| **Jaccard index (IoU)**    | Overlap between successive masks             | Values close to 1 mean masks are similar (algorithm stabilized). Low values mean big changes between iterations. |

Example Discussion (for your report or notebook)

After running GrabCut for 1, 3, and 5 iterations, both qualitative and quantitative differences were observed.
Qualitatively, the segmentation boundaries became more precise, and background artifacts were reduced as iterations increased.
Quantitatively, the foreground fraction decreased slightly, and the Jaccard similarity between iteration 3 and 5 masks approached 0.97, indicating that the algorithm had converged.
Hence, while early iterations (13) significantly improved segmentation, further iterations offered marginal benefit.
